For discussion data.

```python
def add_value(dict_obj, key, value):
    ''' Adds a key-value pair to the dictionary.
        If the key already exists in the dictionary, 
        it will associate multiple values with that 
        key instead of overwritting its value'''
    if key not in dict_obj:
        dict_obj[key] = value
    elif isinstance(dict_obj[key], list):
        dict_obj[key].append(value)
    else:
        dict_obj[key] = [dict_obj[key], value]
```

Find all commit messages and descriptions for all files

```python
files_path = glob.glob('C:/Users/GNR/Desktop/data/git_repos/crop_data/data/discussion/*/*/*.txt')
commit_messages={}
descriptions={}
all_comments = {}
all_authors = {}
all_dates = {}

for file_path in files_path:
    comments = {}
    comments_limits=[]
    authors = {}
    dates = {}
    file = open(file_path, encoding='utf8')
# Checks for commit messages
    for x, line in enumerate(file):
        if line[:12] == "First commit":
            commit_message_start = x
        if line[:10] == "Change-Id:":
            commit_message_end = x
            break
    file.close()
# Checks for description
    file = open(file_path, encoding='utf8')
    for x, line in enumerate(file):
# If it has a commit message
        if line[:12] == "First commit":
            description_end = x
            break
        if line[:10] == "Change-Id:":
            description_end = x
            break
    file.close()
    file = open(file_path, encoding='utf8')
    
    # Check for Comments
    file = open(file_path, encoding='utf8')
    for x, line in enumerate(file):
        if line[:5] == "date:":
            start = x
        if line[:10] == "----------":
            end = x
            tup=(start,end)
            comments_limits.append(tup)
    file.close()
    
    comment = ""
    author = ""
    date = ""
    commit_message = ""
    description = ""
    # Saves all commit messages and descriptions
    file = open(file_path, encoding='utf8')
    for x, line in enumerate(file):
        if x > commit_message_start and x < commit_message_end:
            commit_message+=line
        if x > 1 and x < description_end:
            description+=line
    add_value(commit_messages, ntpath.basename(file_path)[:10], commit_message.replace("\n",""))
    add_value(descriptions, ntpath.basename(file_path)[:10], description.replace("\n",""))
    file.close()
    
    # Saving the comments, author and dates of the comments
    for tuple in comments_limits:
        file = open(file_path, encoding='utf8')
        for x, line in enumerate(file):
            if x > tuple[0] and x < tuple[1]:
                comment+=line
            if x == tuple[0]-1:
                author+=line[8:]
            if x == tuple[0]:
                date+=line[6:]
        add_value(comments, ntpath.basename(file_path)[:10], comment.replace("\n",""))
        add_value(authors, ntpath.basename(file_path)[:10], author.replace("\n",""))
        add_value(dates, ntpath.basename(file_path)[:10], date.replace("\n",""))
        comment = ""
        author = ""
        date = ""
        file.close()
    all_comments.update(comments)
    all_authors.update(authors)
    all_dates.update(dates)

    file.close()
```

Convert lists to dataframes

```python
all_comments_df = pd.Series(all_comments, name='Comment').rename_axis('PR_number').explode().reset_index()
all_authors_df = pd.Series(all_authors, name='Author').rename_axis('PR_number').explode().reset_index()
all_dates_df = pd.Series(all_dates, name='Date').rename_axis('PR_number').explode().reset_index()
commit_messages_df = pd.Series(commit_messages, name='Commit_message').rename_axis('PR_number').explode().reset_index()
descriptions_df = pd.Series(descriptions, name='Description').rename_axis('PR_number').explode().reset_index()
```

```python
commit_messages_df
```

Merge all dataframes to one

```python
all_discussion_specs = pd.merge(all_comments_df, all_authors_df, right_index = True, left_index = True)
del all_discussion_specs['PR_number_y']
```

```python
all_discussion_specs = pd.merge(all_discussion_specs, all_dates_df, right_index = True, left_index = True)
del all_discussion_specs['PR_number_x']
all_discussion_specs
```

```python
commits_and_descriptions = commit_messages_df.merge(descriptions_df, on='PR_number')
```

```python
all_discussion_specs = all_discussion_specs.merge(commits_and_descriptions, on='PR_number', how='outer')
```
